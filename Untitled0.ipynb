```python
# --- Imports ---
import numpy as np
from PIL import Image
from matplotlib import transforms, pyplot as plt
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.metrics import accuracy_score
import torch.nn.functional as F

%matplotlib inline
plt.rcParams['font.family'] = 'serif'
plt.rcParams['axes.linewidth'] = 0.5
GRAY1, GRAY4, GRAY7 = '#231F20', '#646369', '#929497'

# --- Load Data ---
img = np.array(Image.open('zip_numerals.bmp')).astype(int)
X = np.zeros(shape=(480, 16, 16, 1))
y = np.zeros(shape=(480, 1))
example_idx = 0
for i in range(10):
    for j in range(12):
        si, fi = i*17, i*17+16
        sj, fj = j*14, j*14+13
        patch = 1-img[si:fi, sj:fj]
        for k in range(4):
            X[example_idx, 0:16, k:k+13, 0] = patch
            y[example_idx, 0] = i
            example_idx += 1

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=320, random_state=10, stratify=y)
ohe = OneHotEncoder(categories='auto').fit(y)
Y_train = ohe.transform(y_train).toarray()
Y_test = ohe.transform(y_test).toarray()

# --- Visualize Digits ---
w = 20
result = np.zeros(shape=(4+w*5, 4+w*10))
for d in range(10):
    d_idx = np.squeeze(np.argwhere(y_train == d))[:, 0]
    d_idx = np.random.choice(d_idx, 5, replace=False)
    for i, idx in enumerate(d_idx):
        result[i*w+4:i*w+20, d*w+4:d*w+20] = -X_train[idx].reshape((16, 16))
fig, ax = plt.subplots(figsize=(5.5, 3.3), dpi=110)
ax.imshow(result, cmap="gray")
ax.set_aspect('equal', 'datalim')
ax.tick_params(bottom=False, left=False, labelleft=False, labelbottom=False)
for spine in ax.spines.values():
    spine.set_visible(False)
plt.show()

# --- Setup Tensors ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print('Using device:', device)

X_train = torch.tensor(X_train, dtype=torch.float32).permute(0, 3, 1, 2).to(device)
X_test = torch.tensor(X_test, dtype=torch.float32).permute(0, 3, 1, 2).to(device)
Y_train = torch.tensor(Y_train, dtype=torch.float32).to(device)
Y_test = torch.tensor(Y_test, dtype=torch.float32).to(device)

# --- Training Helper ---
def train_model(model, criterion, optimizer, n_epochs, X_train, Y_train, X_test, Y_test, batch_size=32):
    accuracy_test = []
    for epoch in range(n_epochs):
        model.train()
        permutation = torch.randperm(X_train.size(0))
        for i in range(0, X_train.size(0), batch_size):
            indices = permutation[i:i+batch_size]
            batch_x, batch_y = X_train[indices], Y_train[indices]
            optimizer.zero_grad()
            output = model(batch_x)
            loss = criterion(output, batch_y)
            loss.backward()
            optimizer.step()
        model.eval()
        with torch.no_grad():
            y_test_hat = torch.argmax(model(X_test), dim=1)
            y_test_labels = torch.argmax(Y_test, dim=1)
            accuracy = accuracy_score(y_test_labels.cpu(), y_test_hat.cpu())
            accuracy_test.append(accuracy)
    return accuracy_test

# --- Define Net-1 ---
class Net1(nn.Module):
    def __init__(self):
        super(Net1, self).__init__()
        self.fc = nn.Linear(256, 10)
        nn.init.xavier_uniform_(self.fc.weight)
    def forward(self, x):
        x = x.flatten(1)
        x = torch.softmax(self.fc(x), dim=1)
        return x

# --- Define dynamic Net-5 (base) ---
class Net5(nn.Module):
    def __init__(self):
        super(Net5, self).__init__()
        self.conv1 = nn.Conv2d(1, 2, 3, 2, 1)
        self.conv2 = nn.Conv2d(2, 4, 5, 1, 0)

        with torch.no_grad():
            dummy = torch.zeros(1, 1, 16, 16)
            dummy = self.conv1(dummy)
            dummy = self.conv2(dummy)
            flatten_size = dummy.flatten(1).size(1)

        self.fc = nn.Linear(flatten_size, 10)
        nn.init.xavier_uniform_(self.conv1.weight)
        nn.init.xavier_uniform_(self.conv2.weight)
        nn.init.xavier_uniform_(self.fc.weight)

    def forward(self, x):
        x = torch.tanh(self.conv1(x))
        x = torch.tanh(self.conv2(x))
        x = x.flatten(1)
        x = torch.softmax(self.fc(x), dim=1)
        return x

# --- Train Net-1 and Net-5 ---
n_epochs = 30
learning_rate = 4e-3

model1 = Net1().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model1.parameters(), lr=learning_rate)
net1_accuracy = train_model(model1, criterion, optimizer, n_epochs, X_train, Y_train, X_test, Y_test)

model5 = Net5().to(device)
optimizer = optim.Adam(model5.parameters(), lr=learning_rate)
net5_accuracy = train_model(model5, criterion, optimizer, n_epochs, X_train, Y_train, X_test, Y_test)

# --- Visualize Net-5 Filters ---
filters = model5.conv1.weight.data.cpu().numpy()
fig, axes = plt.subplots(1, 2, figsize=(6, 3))
for i in range(2):
    ax = axes[i]
    ax.imshow(filters[i, 0], cmap='gray', interpolation='nearest')
    ax.set_title(f'Filter {i+1}')
    ax.axis('off')
plt.tight_layout()
plt.show()

# --- Interpret the Filters ---
print("Filter Interpretations:")
for i in range(2):
    filt = filters[i, 0]
    vertical_diff = np.sum(np.abs(filt[:, :-1] - filt[:, 1:]))
    horizontal_diff = np.sum(np.abs(filt[:-1, :] - filt[1:, :]))
    if vertical_diff > horizontal_diff:
        pattern = "vertical edge detector"
    elif horizontal_diff > vertical_diff:
        pattern = "horizontal edge detector"
    else:
        pattern = "blob or diagonal feature detector"
    print(f"Filter {i+1}: Likely a {pattern} (vertical diff: {vertical_diff:.2f}, horizontal diff: {horizontal_diff:.2f})")


# --- Improved Net5 Variants (dynamic) ---
class Net5_Original(Net5): pass

class Net5_ReLU(Net5):
    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = F.relu(self.conv2(x))
        x = x.flatten(1)
        x = torch.softmax(self.fc(x), dim=1)
        return x

class Net5_ReLU_MoreChannels(nn.Module):
    def __init__(self):
        super(Net5_ReLU_MoreChannels, self).__init__()
        self.conv1 = nn.Conv2d(1, 8, 3, 2, 1)
        self.conv2 = nn.Conv2d(8, 16, 5, 1, 0)

        with torch.no_grad():
            dummy = torch.zeros(1, 1, 16, 16)
            dummy = self.conv1(dummy)
            dummy = self.conv2(dummy)
            flatten_size = dummy.flatten(1).size(1)

        self.fc = nn.Linear(flatten_size, 10)
        nn.init.xavier_uniform_(self.conv1.weight)
        nn.init.xavier_uniform_(self.conv2.weight)
        nn.init.xavier_uniform_(self.fc.weight)

    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = F.relu(self.conv2(x))
        x = x.flatten(1)
        x = torch.softmax(self.fc(x), dim=1)
        return x

class Net5_Deep(nn.Module):
    def __init__(self):
        super(Net5_Deep, self).__init__()
        self.conv1 = nn.Conv2d(1, 8, 3, 2, 1)
        self.conv2 = nn.Conv2d(8, 16, 3, 1, 1)
        self.conv3 = nn.Conv2d(16, 32, 3, 1, 1)

        with torch.no_grad():
            dummy = torch.zeros(1, 1, 16, 16)
            dummy = self.conv1(dummy)
            dummy = self.conv2(dummy)
            dummy = self.conv3(dummy)
            flatten_size = dummy.flatten(1).size(1)

        self.fc = nn.Linear(flatten_size, 10)
        nn.init.xavier_uniform_(self.conv1.weight)
        nn.init.xavier_uniform_(self.conv2.weight)
        nn.init.xavier_uniform_(self.conv3.weight)
        nn.init.xavier_uniform_(self.fc.weight)

    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = F.relu(self.conv2(x))
        x = F.relu(self.conv3(x))
        x = x.flatten(1)
        x = torch.softmax(self.fc(x), dim=1)
        return x

class Net5_Deep_BN(Net5_Deep):
    def __init__(self):
        super(Net5_Deep_BN, self).__init__()
        self.bn1 = nn.BatchNorm2d(8)
        self.bn2 = nn.BatchNorm2d(16)
        self.bn3 = nn.BatchNorm2d(32)
    def forward(self, x):
        x = F.relu(self.bn1(self.conv1(x)))
        x = F.relu(self.bn2(self.conv2(x)))
        x = F.relu(self.bn3(self.conv3(x)))
        x = x.flatten(1)
        x = torch.softmax(self.fc(x), dim=1)
        return x

# --- Net5_Strong (MaxPool + Dropout) ---
class Net5_Strong(nn.Module):
    def __init__(self):
        super(Net5_Strong, self).__init__()
        self.conv1 = nn.Conv2d(1, 8, 3, padding=1)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(8, 16, 3, padding=1)
        self.conv3 = nn.Conv2d(16, 32, 3, padding=1)

        with torch.no_grad():
            dummy = torch.zeros(1, 1, 16, 16)
            dummy = self.pool(F.relu(self.conv1(dummy)))
            dummy = self.pool(F.relu(self.conv2(dummy)))
            dummy = self.pool(F.relu(self.conv3(dummy)))
            flatten_size = dummy.flatten(1).size(1)

        self.fc1 = nn.Linear(flatten_size, 128)
        self.fc2 = nn.Linear(128, 10)
        self.dropout = nn.Dropout(0.5)

    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = self.pool(x)
        x = F.relu(self.conv2(x))
        x = self.pool(x)
        x = F.relu(self.conv3(x))
        x = self.pool(x)
        x = x.flatten(1)
        x = self.dropout(x)
        x = F.relu(self.fc1(x))
        x = torch.softmax(self.fc2(x), dim=1)
        return x

# --- Train Trials ---
trial_models = [
    ('Trial 1: Net5 Original', Net5_Original()),
    ('Trial 2: Net5 ReLU', Net5_ReLU()),
    ('Trial 3: Net5 ReLU More Channels', Net5_ReLU_MoreChannels()),
    ('Trial 4: Net5 Deep', Net5_Deep()),
    ('Trial 5: Net5 Deep + BN', Net5_Deep_BN()),
    ('Trial 6: Net5 Strong (MaxPool+Dropout)', Net5_Strong())
]

n_epochs_improve = 50
learning_rate_improve = 2e-3
results_improved = {}

for name, model in trial_models:
    model = model.to(device)
    optimizer = optim.Adam(model.parameters(), lr=learning_rate_improve)
    criterion = nn.CrossEntropyLoss()
    print(f"Training {name}...")
    acc = train_model(model, criterion, optimizer, n_epochs_improve, X_train, Y_train, X_test, Y_test)
    results_improved[name] = acc

# --- Final Plot and Results ---
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(range(30), [100*a for a in net1_accuracy], label='Net-1', linestyle='--')
ax.plot(range(30), [100*a for a in net5_accuracy], label='Net-5', linestyle='--')
for name, acc in results_improved.items():
    ax.plot(range(50), [100*a for a in acc], label=name)
ax.set_xlabel('Epoch')
ax.set_ylabel('Test Accuracy (%)')
ax.set_title('Comparison of Net-5 Variants')
ax.legend()
plt.show()

print("\nBest Test Accuracies:")
print(f"Net-1: {max(net1_accuracy)*100:.2f}%")
print(f"Net-5: {max(net5_accuracy)*100:.2f}%")
for name, acc in results_improved.items():
    print(f"{name}: {max(acc)*100:.2f}%")
```
